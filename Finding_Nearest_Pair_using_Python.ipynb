## Idea

Perform Haversine Distance formula Analysis between hotel(Restaurants) locations and nearby_point in Malaysia. The dataset consists of two separate files. The first one is Hotel locations and the second is nearby_point. We want to determine which hotel is near to any given member in the second dataset

```python
import pandas as pd
```

```python
#Replicating Haversine Distance fromula in Python :

from math import radians, cos, sin, asin, sqrt
def dist(lat1, long1, lat2, long2):
    """
Replicating the same formula as mentioned in Wiki
    """
    # convert decimal degrees to radians 
    lat1, long1, lat2, long2 = map(radians, [lat1, long1, lat2, long2])
    # haversine formula 
    dlon = long2 - long1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
```

```python
#Reading the Data from CSV File

main_point = pd.read_csv("/content/pusat_loc.csv")
main_point.head()
```

```python
nearby_point = pd.read_csv("/content/hotspot_loc.csv")
nearby_point.head()
```

```python
#Haversine Equation, to find the distance

from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
```

```python
#Using idxmin you can find the closest Hotel name using,

def find_nearest(lat, long):
    distances = main_point.apply(
        lambda row: dist(lat, long, row['lat'], row['lon']), 
        axis=1)
    return main_point.loc[distances.idxmin(), 'Lokasi']
```

```python
nearby_point['Lokasi'] = nearby_point.apply(
    lambda row: find_nearest(row['lat'], row['lon']), 
    axis=1)

# To check the data frame if it has a new column of hotel name (for each and every member's location in the list)
nearby_point.head()
```

```python
#PART B — Finding the “distance” between member location to the closest hotel

nearby_point = pd.merge(nearby_point,main_point[['Lokasi', 'dun','lat','lon']],on='Lokasi', how='left')
# Rename the new columns as both the columns has same name, and python gets confused 
nearby_point=nearby_point.rename(columns = {'lat_x':'m_lat','lon_x':'m_lon','lat_y':'h_lat','lon_y':'h_lon'})
nearby_point.head()
```

```python
# Creating a new column to generate the output by passing lat long information to Haversine Equation
nearby_point['distance'] = [haversine(nearby_point.m_lon[i],nearby_point.m_lat[i],nearby_point.h_lon[i],nearby_point.h_lat[i]) for i in range(len(nearby_point))]
nearby_point['distance'] = nearby_point['distance'].round(decimals=3)

# Printing the data table 
nearby_point.head()
```

```python
nearby_point['status'] = nearby_point['dun_x'].isin(nearby_point['dun_y'])
```
